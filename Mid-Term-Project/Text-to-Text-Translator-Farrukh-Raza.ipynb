```python
# Importing necessary libraries for User Interface in Translation

from tkinter import *
from tkinter import ttk
from googletrans import Translator , LANGUAGES
from tkinter import messagebox
```

```python
# Createing User Interface widget using the 'tkinter' library

UI = Tk()  # Initializing the UI (UserInterface)
UI.geometry('1080x400')
UI.resizable(0,0)
UI.title("Text Translator for NED Academy - by Farrukh Raza")
UI.config(bg = 'darkolivegreen3')

# Label for Headings
Label(UI, text = "Text to Text Translator", font="arial 20 bold", fg='White', bg='darkolivegreen4', width='20').pack()
Label(UI,text ="Text Translator for NED Academy by Farrukh Raza", font = 'arial 20 bold', fg='White', bg='darkolivegreen4', width='45').pack(side = 'bottom')
```

```python
# Input Text widget
Label(UI,text ="Input Text", font = 'arial 13 bold', bg ='white smoke').place(x=20,y=85)
Input_Text = Text(UI,font = 'arial 10', height = 11, wrap = WORD, padx=5, pady=5, width = 60)
Input_Text.place(x=25,y = 115)

# Output Text widget
Label(UI,text ="Output Text", font = 'arial 13 bold', bg ='white smoke').place(x=600,y=85)
Output_Text = Text(UI, font = 'arial 10', height = 11, wrap = WORD, padx=5, pady= 5, width =60)
Output_Text.place(x = 605 , y = 115)
```

```python

# List of all languages from the 'googletrans' library
language = list(LANGUAGES.values())

# Combobox for Input Language
Input_Lang = ttk.Combobox(UI, values= language, width =22)
Input_Lang.place(x=20,y=60)
Input_Lang.set('Choose Input Language')

# Combobox for Output Language
Output_Lang = ttk.Combobox(UI, values= language, width =22)
Output_Lang.place(x=600,y=60)
Output_Lang.set('Choose Output Language')


def validate(): # This function checks if user has provided all necessary information
    
    if Input_Lang.get() == "Choose Input Language": # Checking if Input Language is selected
        messagebox.showinfo("Error", "Please choose Input Language.")
        
    elif Output_Lang.get() == "Choose Output Language": # Checking if Output Language is selected
        messagebox.showinfo("Error", "Please choose Output Language.")
        
    elif Input_Text.get(1.0, END)=="\n":  # Checking if Input Text is provided
        messagebox.showinfo("Error", "Please provide Input Text.")
        
    else: # If everything is okay then execute the 'Translate_Text' function
        Translate_Text()

        

def Translate_Text(): # This function translates the provided text in the chosen Output Language
    translator = Translator()
    translated=translator.translate(text= Input_Text.get(1.0, END) , src=Input_Lang.get(), dest=Output_Lang.get())
    Output_Text.delete(1.0, END)
    Output_Text.insert(END, translated.text)


# Creating Button to execute the translation process
Translate_Button = Button(UI, text = 'Translate', font='arial 12 bold', fg='white', pady=5, command=validate, bg='darkolivegreen4', activebackground='darkolivegreen', width='8')
Translate_Button.place(x = 485, y = 180)



# Open the User Interface in a new Window
UI.mainloop()
```
